```python
import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials
```

```python
companies = open("companies.txt").read().splitlines()

stocks_df = pd.DataFrame()

for company in companies:
    ticker = yf.Ticker(company)

    stock_df = ticker.history(start='2000-01-01')

    stock_df['Company'] = company

    stocks_df = stocks_df.append(stock_df)

    stock_df['Close'].plot()
```

```python
# Save stocks df to csv
stocks_df.to_csv(f'stock-price/stock-prices.csv', index=False) 
```
